```python
import pandas as pd

url = 'https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv'
df = pd.read_csv(url)

print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
print(df.head(5))
```

```python
import random

seed1 = 4721
seed2 = 3803
seed3 = 9156
seed4 = 2047

sample1 = df.sample(n=1000, random_state=seed1).reset_index(drop=True)
sample2 = df.sample(n=1000, random_state=seed2).reset_index(drop=True)
sample3 = df.sample(n=1000, random_state=seed3).reset_index(drop=True)
sample4 = df.sample(n=1000, random_state=seed4).reset_index(drop=True)

print("Sample 1 shape:", sample1.shape)
print("Sample 2 shape:", sample2.shape)
print("Sample 3 shape:", sample3.shape)
print("Sample 4 shape:", sample4.shape)

for i, s in enumerate([sample1, sample2, sample3, sample4], 1):
    pos = s['Positive'].sum()
    neg = 1000 - pos
    print(f"Sample {i} — Positive: {pos}, Negative: {neg}")
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('vader_lexicon')
nltk.download('punkt_tab')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.metrics import confusion_matrix, classification_report

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    processed_text = ' '.join(lemmatized_tokens)
    return processed_text

analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    sentiment = 1 if scores['pos'] > 0 else 0
    return sentiment

samples = [sample1, sample2, sample3, sample4]

for i, sample in enumerate(samples, 1):
    sample['reviewText'] = sample['reviewText'].apply(preprocess_text)
    sample['sentiment'] = sample['reviewText'].apply(get_sentiment)
    print(f"\n===== Sample {i} =====")
    print("\nConfusion Matrix:")
    print(confusion_matrix(sample['Positive'], sample['sentiment']))
    print("\nClassification Report:")
    print(classification_report(sample['Positive'], sample['sentiment']))
```
